# Literature Review

Considerable scholarly attention has been given to Trumpâ€™s anti-immigrant rhetoric as a candidate and president. The frequency and tone of immigration topics in Trumpâ€™s public speeches were studied to determine if his rhetoric influenced public opinion or Congressional legislation.<sup>1</sup> His speech patterns were analyzed to show the use of metaphors that simultaneously denigrate immigrants and appeal to his base.<sup>2</sup> His supporters were found to use language patterns that differed from those of Trump detractors in politically charged Facebook posts.<sup>3</sup> They conveyed the superiority of Trump supporters and the American people compared to the inferiority of Muslims, hinting at the political ideology enveloped in the language.<sup>3</sup> Sagir and Mockabee conclude that Trump did not actively win over the audience with his anti-immigrant rhetoric, but rather voiced pre-existing feelings that agreed with the Republican base.<sup>4</sup>

Although Eshbaugh-Soha and Barnes claim that Trumpâ€™s anti-immigration was all bluster as it did not push Congress to pass sweeping immigration legislation and did not alter the publicâ€™s opinion on the issue,<sup>1</sup> Hamlin argues that there are long-lasting effects of Trumpâ€™s changes to existing immigration systems.<sup>5</sup> The changes include normalizing building a wall to deter unauthorized immigration, and denying visas, limiting asylum grants, and packing immigration court benches to reduce legal immigration.<sup>5</sup> The powers of enforcement agencies such as USCIS and ICE were expanded, and aggressive tactics were encouraged.<sup>5</sup> There was also a surge in naturalization rates called â€œdefensive citizenship,â€ where immigrants raised their residency status against increased scrutiny and uncertainty.<sup>5</sup>  Despite Trumpâ€™s extremist approach, it has put Democrats on the defensive around issues of immigration and border security and Biden has not deviated far from Trumpâ€™s policies.<sup>5</sup> Trumpâ€™s legacy regarding US immigration will endure for years to come, even if we do not yet know the full scale of the impact.

To date, scholars have researched the nuances of Trumpâ€™s speech, analyzed his public policy proposals, and examined the sentiment of public reactions online. However, little attention has been given to the quantitative results of Trumpâ€™s rhetoric and public policy. Did incoming immigration decrease, and deportations increase during his presidency?

# Methods

To investigate the quantitative changes during Trump's presidency, three datasets from the Migration Policy Institute and the Department of Homeland Security were used to examine the numbers of new Legal Permanent Residents (LPR), Naturalized US Citizens, Removals (deportations), and (voluntary) Returns each year. Each dataset contains data from 1999 to 2019 or 2021. The data was enhanced to include the sitting president for each year, a dummy variable to indicate the party of the president, a variable to indicate presidential Term 1 or Term 2, a variable 1-4 for the year of the presidential term, and a dummy variable to indicate whether the president was Trump. From this data and the literature review, four hypotheses were formed:

H<sub>1</sub>: Legal Permanents Residents (LPR) entering the US from 2017 to 2020 will decrease compared to previous presidential terms.

H<sub>2</sub>: Naturalized US Citizens will increase from 2017 to 2020 compared to previous presidential terms.

H<sub>3</sub>: Removals (deportations) will increase from 2017 to 2020 compared to previous presidential terms.

H<sub>4</sub>: Returns will increase from 2017 to 2020 compared to previous presidential terms.

Time series graphs of each of the variables of interest were produced to show trends. To provide a more accurate comparison, Clinton's and Biden's partial data are excluded as well as data from 2020 due to the pandemic. The averages of the presidential termâ€™s first three years are compared from Bushâ€™s first term to Trump. A difference-in-means test was used to determine if the variable of interest was significantly different during Trumpâ€™s presidency from preceding presidential terms. After completing the t-test to examine the difference of means, a multivariate linear regression analysis was constructed to determine if any relationship exists between the variable of interest and Trumpâ€™s presidency.
